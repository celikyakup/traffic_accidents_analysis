```python
plt.figure(figsize=(20,10))
sns.countplot(x="Speed_limit",hue="Accident_Severity",data=df)
plt.title("The severity of the accident according to the speed limit")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Weather_Conditions",hue="Accident_Severity",data=df)
plt.title("accident severity according to weather conditions")
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
plt.show()
```

```python
sns.relplot(data=df,x="Longitude",y="Latitude")
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Day_of_Week",data=df)
plt.title("Number of Accident by Days of week")
plt.xlabel("Days of week")
plt.ylabel("Number of Accident")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Year",data=df)
plt.title("Number of Accident by years")
plt.xlabel("Year")
plt.ylabel("Number of Accident")
plt.show()
```

```python
plt.figure(figsize=(22,10))
sns.countplot(x="Road_Surface_Conditions",hue="Year",data=df)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Road_Type",data=df,hue="Accident_Severity")
plt.title("Accident Severity by road type")
plt.xlabel("Road Type")
plt.ylabel("Number of Accident")
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
plt.show()
```
